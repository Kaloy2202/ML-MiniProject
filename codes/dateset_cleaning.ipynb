```python
#load the dataset
file_path = "/content/cleaned_dataset/ML/processed_pond1.csv"
data = pd.read_csv(file_path)
```

```python
# # Drop unnecessary attributes
columns_to_drop = ['entry_id']
data = data.drop(columns=columns_to_drop)
```

```python
# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())
print(data.info())
```

```python
 # Handling missing values
 # Fill missing values in Ammonia with the mean
data['Ammonia(g/ml)'] = data['Ammonia(g/ml)'].fillna(data['Ammonia(g/ml)'].mean())
# data['Temperature (C)'] = data['Temperature (C)'].fillna(data['Temperature (C)'].mean())
# data['Turbidity(NTU)'] = data['Turbidity(NTU)'].fillna(data['Turbidity(NTU)'].mean())
# data['Dissolved Oxygen (mg/L)'] = data['Dissolved Oxygen (mg/L)'].fillna(data['Dissolved Oxygen (mg/L)'].mean())
# data['PH'] = data['PH'].fillna(data['PH'].mean())
# data['Nitrate (mg/L)'] = data['Nitrate (mg/L)'].fillna(data['Nitrate (mg/L)'].mean())
# data = data.dropna(subset=['Unnamed: 0'])
# # Interpolate missing values in Fish_Length and Fish_Weight
# data['Total_length (cm)'] = data['Total_length (cm)'].interpolate(method='linear')
# data['Weight (g)'] = data['Weight (g)'].interpolate(method='linear')

# Verify that there are no missing values left
print("\nMissing Values After Handling:")
print(data.isnull().sum())
```

```python
# Ensure 'created_at' is in datetime format
data['created_at'] = pd.to_datetime(data['created_at'], errors='coerce')

# Check for any invalid timestamps
invalid_timestamps = data['created_at'].isnull()
# if invalid_timestamps > 0:
#     print(f"\nFound {invalid_timestamps} invalid timestamps. Dropping these rows.")
#     data = data.dropna(subset=['created_at'])

# Verify the format
print("\nDataset after Timestamp Formatting:")
print(data['created_at'].head())
data.info()
```

```python

# Ensure 'created_at' is in datetime format
data['created_at'] = pd.to_datetime(data['created_at'])

data['Fish_length(cm)'] = pd.to_numeric(data['Fish_length(cm)'], errors='coerce')

# Assign each timestamp to a bi-weekly period
data['bi_weekly_period'] = data['created_at'].dt.to_period('2W')

data.info()
```

```python
# List of water quality parameters
water_quality_features = ['Temperature(C)', 'Turbidity(NTU)', 'Dissolved Oxygen(g/ml)',
                           'PH', 'Ammonia(g/ml)', 'Nitrate(g/ml)', 'Fish_length(cm)', 'Fish_weight(g)']

# Group by bi-weekly period and compute mean for each period
aggregated_quality_pond = data.groupby('bi_weekly_period')[water_quality_features].mean().reset_index()
aggregated_quality_pond.info()
```

```python
aggregated_quality_pond.head()
```

```python
aggregated_quality_pond.sort_values(by='bi_weekly_period', inplace=True)
```

```python
# Calculate growth rate (g/2-weeks)
aggregated_quality_pond['Growth_Rate(g/2weeks)'] = aggregated_quality_pond['Fish_weight(g)'].diff()

#change nan to 0 for the first row
aggregated_quality_pond['Growth_Rate(g/2weeks)'] = aggregated_quality_pond['Growth_Rate(g/2weeks)'].fillna(0)
```
